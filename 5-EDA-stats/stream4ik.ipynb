```python
# функция из прошлого проекта для знакомства с данными
def df_description(df):
    print(f'В таблице {df.shape[0]} строк и {df.shape[1]} столбцов.\n')
    print('Наименование столбцов таблицы:')
    display(df.columns)
    print('Первые 10 строк таблицы:')
    display(df.head(10))
    print('Типы данных категорий и число значений в них:\n')
    df.info()


# фунцкия описывающая выборки
def sample_comparison(df, selection, col):
    return pd.DataFrame(
        {'Описание выборки': df.loc[selection, col].describe(),
         'Остальные данные': df.loc[~selection, col].describe()})


# функция строит матрицу рассеяния для продажами игр у платформы и оценками и считает корреляцию
def platform_relations(platform):
    for col, name in {'critic_score': 'оценками критиков', 'user_score': 'оценками игроков'}.items():
        sns.scatterplot(data=work_set[work_set.platform == platform],
                        x="world_sales", y=col)

        plt.title(f'Зависимость между продажами игр у {platform} и {name}', fontsize=14)
        plt.xlabel('Суммарные продажи игры')
        plt.ylabel('Оценка игры')
        plt.show()

        print('Корреляция между метриками:',
              work_set[work_set.platform == platform].world_sales.corr(work_set.loc[work_set.platform == platform, col]))


# функция находит топ 5 продаж по таргету и выводит столбчатый график
# следующим шагом выводит по этим же целям диаграммы размаха
def region_explorer(df, region, target):
    table = df.groupby(target[0])[region].sum().sort_values(ascending=False).head().reset_index()
    region_name = region.split('_')[0].upper()

    sns.barplot(data=table, x=target[0], y=region)
    plt.title(f'Топ-5 {target[2]} региона {region_name} по продажам', fontsize=14)
    plt.xlabel(target[1])
    plt.ylabel('Общие продажи')
    plt.show()

    top_set = df[df[target[0]].isin(list(table[target[0]]))]

    # ограничим график 2.6 * значение 3 квантиля (1 межквартильный размах для ящика + 1.5 для уса + 0.1 для запаса)
    y_max = df.groupby(target[0])[region].quantile(0.75).max() * 2.6 

    sns.boxplot(data=top_set, x=target[0], y=region, order=list(table[target[0]]))
    plt.ylim(-0.01, y_max)
    plt.title(f'Распределения продаж игр топ-5 {target[2]} региона {region_name}', fontsize=14)
    plt.xlabel(target[1])
    plt.ylabel('Продажи игр')
    plt.show()
    
    
# то же для рейтинга ESRB
def rating_explorer(df, region):
    target = ['rating', 'Рейтинг']
    table = df.groupby(target[0])[region].sum().sort_values(ascending=False).reset_index()
    region_name = region.split('_')[0].upper()

    sns.barplot(data=table, x=target[0], y=region)
    plt.title(f'Продажи игр в зависимости от рейтинга ESRB региона {region_name}', fontsize=14)
    plt.xlabel(target[1])
    plt.ylabel('Общие продажи')
    plt.show()

    top_set = df[df[target[0]].isin(list(table[target[0]]))]

    # ограничим график 2.6 * значение 3 квантиля (1 межквартильный размах для ящика + 1.5 для уса + 0.1 для запаса)
    y_max = df.groupby(target[0])[region].quantile(0.75).max() * 2.6 

    sns.boxplot(data=top_set, x=target[0], y=region, order=list(table[target[0]]))
    plt.ylim(-0.01, y_max)
    plt.title(f'Распределения продаж игр среди рейтинга ESRB региона {region_name}', fontsize=14)
    plt.xlabel(target[1])
    plt.ylabel('Продажи игр')
    plt.show()
    
    
# функция для проверки гипотезы равенства двух независимых и неравных выборок
def hypothesting(df_1, df_2):
    # уровень статистической значимости
    alpha = 0.05
    
    results = st.ttest_ind(df_1, df_2, equal_var = False)
    
    print('p-значение: ', results.pvalue)
    print('статистика разности:', results.statistic)
    if results.pvalue < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

Теперь мы можем спрогнозировать потенциально популярный продукт и спланировать рекламные кампании на 2017 год для компании, которая продает игры по всему миру.

**В ходе исследования мы пришли к следующим выводам:**
1. **Игровой рынок динамичен.** Он молод, постоянно меняется и находится на стадии становления. Последние изменения происходили в 2008-2012 годах - в связи с появлением мобильных устройств и мобильных, рынок разделился на 2 части. Предыдущие изменения произошли в 1993 году в связи с началом повсеместного распространения персональных ЭВМ. Таким образом можно полагать, что каждые 15-20 лет на рынке происходит сдвиг формации, т.к. сейчас мы находимся в середине цикла, то можно уверенно исследовать ситуацию и формировать прогнозы.


2. **Игровые консоли следуют поколениями.** В связи с появлением новых технологий и совершенствованием старых игровые консоли имеют ограниченный жизненный цикл - 11 лет, причем разные поколения пересекаются и могут сосуществовать до 5 лет. Популяренейшие консоли последнего поколения - PlayStation 4, Xbox One, Nintendo 3DS.


3. **PlayStation 4.** Распространена по всему миру, наиболее популярна в Европе. Подходящий выбор, если делать ставку на универсальность и продвижение бренда.


4. **Xbox One.** Лидер Северной Америки, популярна в Европе. Демонстрирует лучшие доходы с игры среди остальных платформ. Если ориентироваться на западный рынок и прибыль, то это лучший вариант.


5. **Nintendo 3DS.** Портативная консоль, лидер Японского рынка. В последнее время есть проблемы с ростом продаж, поэтому Nintendo в октябре 2016 года анонсировала консоль Nintendo Switch гибридного типа, чтобы охватить рынки портативных игровых консолей и стационарных. Так как новая консоль еще не вышла, а прошлая устаревает, то на данный момент не рекомендуется для коммерческой экспансии.


6. **Шутеры.** однозначный лидер среди жанров по прибыльности с игры. Популярен на Западе, но не популярен в Японии. Прекрасный выбор для экспансии на западном рынке.


7. **Спортивные игры.** второй жанр по прибыльности с игры. Достаточно распространен на Западе, но не популярен в Японии.


8. **Экшены.** посредственная прибыльность, но поразительная распространенность. Вероятно это связано с простотой произоводства и низкой себестоимостью. Популярен по всему миру, но в Японии уступает играм жанра РПГ. Универсальный вариант.


9. **Оценки критиков.** Оценки критиков не гарантируют высокие продажи, но при этом являются необходимым условием для них. Возможно покупатели не осознают, что держат игру высоко оцененную критиками, поэтому им надо подсказать, что их выбор, очевидно, правилен. Например демонстрацией оценок критиков в рекламных компаниях, на постерах и обложках.


10. **Рейтинговая система.** Игры взрослого рейтинга пользуются безусловной популярностью на западном рынке, в Японии наиболее популярны игры для подростков. Также рекомендуется демонстрировать подходящий рейтинг в рекламных компаниях.
